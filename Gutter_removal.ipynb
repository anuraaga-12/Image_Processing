```python
import cv2 as cv
import numpy as np
from google.colab.patches import cv2_imshow
#import necessary libraries
```

```python
serial=input('Enter the image number:\n')
#taking input to set the image
```

```python
img=cv.imread('/content/gutters1.JPG')
if(serial=='2'):
    img=cv.imread('/content/gutters2.JPG')
if(serial=='3'):
    img=cv.imread('/content/gutters3.JPG')
cv2_imshow(img)
#setting image to be cleaned through input
#image is set to gutter1 by default
```

```python
img=cv.cvtColor(img,cv.COLOR_BGR2GRAY)
dilated_img=cv.dilate(img,np.ones((5,5),np.uint8))
mblur_img=cv.medianBlur(dilated_img,9)
cv2_imshow(dilated_img)
cv2_imshow(mblur_img)
#dilating and blurring the image to obtain the backgorund
```

```python
diff_img=cv.absdiff(img,mblur_img)
cv2_imshow(diff_img)
diff_img=255-diff_img
cv2_imshow(diff_img)
#taking diff between foreground and background to obtain letters and inverting the colors
```

```python
norm_img=diff_img.copy()
cv.normalize(diff_img,norm_img,alpha=0,beta=255,norm_type=cv.NORM_MINMAX,dtype=cv.CV_8UC1)
cv2_imshow(norm_img)
#normalizing the image
```

```python
 adaptive_thersold_img =255 - cv.adaptiveThreshold(norm_img,255,cv.ADAPTIVE_THRESH_MEAN_C,cv.THRESH_BINARY_INV,21,21)
```

```python
cv2_imshow( adaptive_thersold_img)
cv.waitKey(0)
#outputing the cleaned page
```
