```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
#importing necessary libraries
```

```python
serial=input('Enter the image number:\n')
#taking input to set the image
```

```python
img=cv.imread('/content/cctv1.JPG')
if(serial=='2'):
    img=cv.imread('/content/cctv2.JPG')
if(serial=='3'):
    img=cv.imread('/content/cctv3.JPG')
if(serial=='4'):
    img=cv.imread('/content/cctv4.JPG')
cv2_imshow(img)
#setting image to be cleaned through input
#image is set to cctv1 by default
```

```python
gray=cv.cvtColor(img,cv.COLOR_BGR2GRAY)
cv2_imshow(gray)
#grayscaling image
```

```python
eq_gray=cv.equalizeHist(gray)
cv2_imshow(eq_gray)
#equlaizing the image
```

```python
clahe=cv.createCLAHE(clipLimit=40)
clahe_eq_gray=clahe.apply(eq_gray)
cv2_imshow(clahe_eq_gray)
#applying CLAHE
```

```python
ret,thresh=cv.threshold(clahe_eq_gray,50,255,cv.THRESH_TOZERO)
#thresholding the image
```

```python
cv2_imshow(thresh)
cv.waitKey(0)
#outputting enhanced image of cctv1
```
